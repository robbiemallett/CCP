# Now do the retracking

```python
ranges_to_128 = np.array(d['window_del_20_ku'])*3e8/2


bin_time = 0.2342/3e8

waveforms = d['pwr_waveform_20_ku']
alts = d['alt_20_ku']
```

## Calculate leading edge widths

```python
leading_edge_widths = np.empty(waveforms.shape[0])

for i in tqdm.trange(waveforms.shape[0]):

    waveform = np.array(waveforms[i])
    bin_70 = np.argmax(waveform>7*np.max(waveform)/10)
    bin_30 = np.argmax(waveform>3*np.max(waveform)/10)
    leading_edge_widths[i] = bin_70 - bin_30
```

## Classify into leads, floes and oceans

```python
specular_echoes =  np.where((np.array(d['stack_std_20_ku'])<6.28)&(np.array(d['stack_peakiness_20_ku'])>18), 1, 0)
diffuse_echoes =  np.where((np.array(d['stack_std_20_ku'])>6.28)&(np.array(d['stack_peakiness_20_ku'])<9), 1, 0)
simple_leading_edges = np.where(leading_edge_widths <= 3, 1, 0)


leads = np.multiply(specular_echoes, simple_leading_edges)
floes = np.multiply(diffuse_echoes, simple_leading_edges)
```

## Retrack floes at 50%

```python
floe_two_way_return_times = np.full(waveforms.shape[0], np.nan)

for i in tqdm.trange(waveforms.shape[0]):
    
    if floes[i]:

        bin_50 = np.argmax(waveform>np.max(waveform)/2)

        time_before_128 = (128-bin_50)*bin_time

        floe_two_way_return_times[i] = d['window_del_20_ku'][i] - time_before_128
        
floe_two_way_return_times[floe_two_way_return_times>1e270] = np.nan

plt.plot(floe_two_way_return_times)
```

## Retrack leads at peak power

```python
lead_two_way_return_times = np.full(waveforms.shape[0], np.nan)

for i in tqdm.trange(waveforms.shape[0]):
    
    if leads[i]:

#         bin_max = np.argmax(waveform)
        
        bin_max = np.argmax(waveform>np.max(waveform)/2)

        time_before_128 = (128-bin_max)*bin_time

        lead_two_way_return_times[i] = d['window_del_20_ku'][i] - time_before_128
        
lead_two_way_return_times[lead_two_way_return_times>1e270] = np.nan

plt.plot(lead_two_way_return_times)
```

```python
plt.plot(floe_two_way_return_times,color='lightblue',label='floe')
plt.plot(lead_two_way_return_times,color='r',label='lead')
plt.ylabel('Two Way Return Time')
plt.xlabel('Waveform Number')
plt.legend()
# plt.ylim(0,1000)

leads_copy = lead_two_way_return_times.copy()
```

```python
nan_args = np.argwhere(np.isnan(lead_two_way_return_times))[:,0]

good_args = np.argwhere(~np.isnan(lead_two_way_return_times))[:,0]

good_vals = lead_two_way_return_times[~np.isnan(lead_two_way_return_times)]

print(nan_args.shape, good_args.shape, good_vals.shape)

leads_copy[np.isnan(leads_copy)]  = np.interp(x=nan_args, xp=good_args, fp=good_vals)
```
